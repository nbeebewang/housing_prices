```python
import pandas as pd
import numpy as np
from mpl_toolkits.basemap import Basemap
from matplotlib.patches import Polygon
import string
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
%matplotlib inline
import math
import os
```

```python
df = pd.read_csv('nyctestscores.csv', delimiter = ",")
```

```python
q1_price = df['mean score'].quantile([0.25, 0.5, 0.75]).values[0]
q2_price = df['mean score'].quantile([0.25, 0.5, 0.75]).values[1]
q3_price = df['mean score'].quantile([0.25, 0.5, 0.75]).values[2]
```

```python
price_brackets = np.empty(len(df))

for i in range(len(df)):
    if df['mean score'][i] < q1_price:
        price_brackets[i] = 1
    elif ((df['mean score'][i] >= q1_price) & (df['mean score'][i] < q2_price)):
        price_brackets[i] = 2
    elif ((df['mean score'][i] >= q2_price) & (df['mean score'][i] < q3_price)):
        price_brackets[i] = 3
    else:
        price_brackets[i] = 4
    
df["price_bracket"] = price_brackets
```

```python
list_1 = df[df['price_bracket'] == 1.0]['district'].values
list_2 = df[df['price_bracket'] == 2.0]['district'].values
list_3 = df[df['price_bracket'] == 3.0]['district'].values
list_4 = df[df['price_bracket'] == 4.0]['district'].values
```

```python
fig = plt.figure(figsize=(15,5))

themap = Basemap(projection='merc',
                llcrnrlon = -74.097444,              # lower-left corner longitude
                llcrnrlat = 40.533572,               # lower-left corner latitude
                urcrnrlon = -73.831712,               # upper-right corner longitude
                urcrnrlat = 40.9,               # upper-right corner latitude
                resolution = 'h',
                area_thresh = 0.1,
              )

themap.drawcoastlines()
themap.drawcountries()
themap.drawstates()
themap.drawrivers()
themap.fillcontinents(color = 'gainsboro')
themap.drawmapboundary(fill_color='steelblue')

themap.readshapefile('./plswork/plswork', 'regions', drawbounds = True)

# collect the names from shapefile attributes so we can look up the shape object by its name
region_names = []
for region_dict in themap.regions_info: 
    region_names.append(region_dict['SchoolDist'])

ax = plt.gca()

# get region and draw the filled polygon 

for elt in list_1: 
    seg = themap.regions[region_names.index(elt)]
    poly = Polygon(seg, facecolor= "#ca0020", edgecolor='#ca0020')
    ax.add_patch(poly)

for elt in list_2: 
    seg = themap.regions[region_names.index(elt)]
    poly = Polygon(seg, facecolor= "#f4a582", edgecolor="#f4a582")
    ax.add_patch(poly)

for elt in list_3: 
    seg = themap.regions[region_names.index(elt)]
    poly = Polygon(seg, facecolor= "#92c5de", edgecolor="#92c5de")
    ax.add_patch(poly)

for elt in list_4: 
    seg = themap.regions[region_names.index(elt)]
    poly = Polygon(seg, facecolor= "#0571b0", edgecolor="#0571b0")
    ax.add_patch(poly)

patch1 = mpatches.Patch(color="#ca0020", label='Less than ' + str(int(q1_price)))
patch2 = mpatches.Patch(color="#f4a582", label='Between ' + str(int(q1_price)) + ' and ' + str(int(q2_price)))
patch3 = mpatches.Patch(color="#92c5de", label='Between ' + str(int(q2_price)) + ' and ' + str(int(q3_price)))
patch4 = mpatches.Patch(color="#0571b0", label='Greater than ' + str(int(q3_price)))

plt.legend(handles=[patch1, patch2, patch3, patch4], loc = 3, prop={'size':6}, title = "Mean Test Scores by School District")

plt.show()
```
